```python
path = "./Sales_Data"
files = [file for file in os.listdir(path)] 

```

```python
all_months9= pd.DataFrame()

for file in files:
    current_data = pd.read_csv(path+"/"+file)
    all_months9 = pd.concat([all_months9, current_data])
    

    
```

# adding month column

```python
null_df = df[df.isna().any(axis=1)]
display(null_df.head())
```

```python
df=df.dropna(how='all')
df
```

```python
df=df[df['Order Date'].str[0:2]!='Or']
```

```python
df['month']=df['Order Date'].str[0:2]
```

```python
df
```

```python
df['month']=df['month'].astype('int32')
```

```python
df
```

```python
df['month'].unique()
```

```python
df['Price Each']=pd.to_numeric(df['Price Each'])
```

```python
df['Quantity Ordered']=df['Quantity Ordered'].astype('int32')
```

```python
df['sales price']=df['Quantity Ordered'] * df['Price Each']
```

```python
df
```

# adding city column

```python
df['city']=df['Purchase Address'].apply(lambda x: x.split(',')[1] +'('+ x.split(',')[2].split(' ')[1]+')')
```

```python
df
```

# Adding hr in a day column

```python
df['Hour'] = pd.to_datetime(df['Order Date']).dt.hour
```
